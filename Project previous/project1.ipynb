```python
# Reading and cleaning weather data

syrWeaather = pd.read_csv("C:/Users/nirmi/Desktop/SU_SEM3/IntroToML/Project/Data sets/weatherData/syracuse_weather_oct_nov_dec2022.csv", parse_dates=['datetime'])
newWeaather = pd.read_csv("C:/Users/nirmi/Desktop/SU_SEM3/IntroToML/Project/Data sets/weatherData/newark_weather_oct_nov_dec2022.csv",parse_dates=['datetime'])
chiWeaather = pd.read_csv("C:/Users/nirmi/Desktop/SU_SEM3/IntroToML/Project/Data sets/weatherData/chicago_weather_oct_nov_dec2022.csv",parse_dates=['datetime'])
wasWeaather = pd.read_csv("C:/Users/nirmi/Desktop/SU_SEM3/IntroToML/Project/Data sets/weatherData/washingtondc_weather_oct_nov_dec2022.csv",parse_dates=['datetime'])
denWeaather = pd.read_csv("C:/Users/nirmi/Desktop/SU_SEM3/IntroToML/Project/Data sets/weatherData/denver_weather_oct_nov_dec2022.csv",parse_dates=['datetime'])

syrWeaather.head()
newWeaather.head()
chiWeaather.head()
wasWeaather.head()
denWeaather.head()
```
