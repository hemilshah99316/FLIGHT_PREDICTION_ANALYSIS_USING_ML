```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
import xgboost as xgb
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
Flight_weather = pd.read_csv(r"C:\Users\VIKAS\Documents\Intro to ML\Project\datasets\Flight_weather.csv", parse_dates =['Date'])
test_Flight_weather = pd.read_csv(r"C:\Users\VIKAS\Documents\Intro to ML\Project\datasets\test_Flight_weather.csv", parse_dates =['Date'])
```

```python
# Feature Engineering 
Flight_weather['date'] = pd.to_datetime(Flight_weather['Date'])
Flight_weather['Day of Week'] = Flight_weather['date'].dt.dayofweek
Flight_weather['Month'] = Flight_weather['date'].dt.month
Flight_weather.drop(['Date', 'date'], axis=1, inplace=True)
Flight_weather
```

```python
# Feature Engineering for test data
test_Flight_weather['date'] = pd.to_datetime(test_Flight_weather['Date'])
test_Flight_weather['Day of Week'] = test_Flight_weather['date'].dt.dayofweek
test_Flight_weather['Month'] = test_Flight_weather['date'].dt.month
test_Flight_weather.drop(['Date', 'date'], axis=1, inplace=True)
test_Flight_weather
```

```python
# Separate the target variable and features
y = Flight_weather['Arr_Delay_Status']
X = Flight_weather.drop(columns=['Arr_Delay_Status'])

X1 = test_Flight_weather.drop(columns=['Arr_Delay_Status'])
y1 = test_Flight_weather['Arr_Delay_Status']
```

```python
# Encode categorical variables using OneHotEncoder and the target variable using LabelEncoder
categorical_features_Flight_weather = X.select_dtypes(include=['object']).columns.tolist()
```

```python
# Encode categorical variables using OneHotEncoder and the target variable using LabelEncoder
categorical_features_test_Flight_weather = X1.select_dtypes(include=['object']).columns.tolist()
```

```python
# Applying OneHotEncoder
X_encoded = pd.get_dummies(X, columns=categorical_features_Flight_weather)
```

```python
# Applying OneHotEncoder
X1_encoded = pd.get_dummies(X1, columns=categorical_features_test_Flight_weather)
```

```python
# Label encode the target variable
label_encoder = LabelEncoder()
y_encoded = label_encoder.fit_transform(y)
```

```python
y1_encoded = label_encoder.fit_transform(y1)
```

```python
# Split the dataset into training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
rf_classifier = RandomForestClassifier(random_state=42)
rf_classifier.fit(X_train, y_train)

y_pred = rf_classifier.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Classification Report: \n{classification_report(y_test, y_pred)}")
```

```python
xgb_classifier = xgb.XGBClassifier(objective='multi:softmax', num_class=3, random_state=42)
xgb_classifier.fit(X_train, y_train)

y_pred = xgb_classifier.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
print(f"Classification Report: \n{classification_report(y_test, y_pred)}")
```
